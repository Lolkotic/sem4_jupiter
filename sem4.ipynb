```python
df['grade'].value_counts()
```

```python
data = df['grade'].value_counts()

plt.figure(figsize=(10, 10))
plt.pie(data, autopct='%1.1f%%')
plt.legend(['Оценка недвижимости - 7', '8', '9','6','10','11','5','12','4','13','3','1']);
```

Больше всего недвижимости со средней оценкой,соотвественно она влияет на цену прямо пропорционально))

```python
sns.jointplot(x=df['price'], y=df['bedrooms'], kind='reg');
```

Чем больше спален в доме, тем выше его стоимость

```python
sns.boxplot(x=df['price'], y=df['bathrooms'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('bathrooms')
plt.title('Distribution of price by bathrooms');
```

Чем больше спален(8) в доме , тем выше его стоимость

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
```

Чем больше квадратура, тем дороже стоимость

```python
sns.boxplot(x=df['price'], y=df['waterfront'].astype('str'), whis=1);
plt.xlabel('price')
plt.ylabel('waterfront')
plt.title('Distribution of price by waterfront');
```

Цена явно больше с хорошим видом)))
